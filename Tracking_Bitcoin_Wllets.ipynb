###### imports

```python
import requests
import pandas as pd
from pprint import pprint
import cryptoapis
import ast
import json
import pickle
from datetime import datetime
```

##### Function to query transactions for a wallet

```python
def get_wallet_transactions(wallet):
    url='https://blockchain.info/rawaddr/{}'.format(wallet)
    print(url)
    resp=requests.get(url)
    return resp.text
```

#### Wallets we want to monitor (including known WannaCry wallet)

```python
WALLETS_TO_MONITOR=set(['12t9YDPgwueZ9NyMgw519p7AA8isjr6SMw','TB2iP6zPR3Z7VAAq8kRnZgsoPP6qa9kjgH'])
```

#### Date for first seen, last seen

```python
datestr=datetime.today().strftime('%Y%m%d')
print(datestr)
```

#### Read in stored wallet info

```python
with open('wallet_info.pkl', 'rb') as f:

    wallet_info = pickle.load(f) # deserialize using load()
```

#### Query for new transactions for each wallet, check for new activity

```python
for wallet in WALLETS_TO_MONITOR:
    print('Examining wallet {}'.format(wallet))
    transactions=get_wallet_transactions(wallet)
    if wallet in wallet_info and len(transactions) > wallet_info[wallet]['number_of_transactions']:
        print('New activity detected on wallet {}!'.format(wallet))
        wallet_info[wallet]={'number_of_transactions':len(transactions),'transactions':transactions, 'last_seen':datestr}
    elif wallet not in wallet_info:
        wallet_info[wallet]={'number_of_transactions':len(transactions),'transactions':transactions, 'last_seeb':datestr,'last_seen':datestr}
#print(wallet_info)
```

#### Save the updated wallet info to disk

```python
with open('wallet_info.pkl', 'wb') as f:
    pickle.dump(wallet_info, f)
```
